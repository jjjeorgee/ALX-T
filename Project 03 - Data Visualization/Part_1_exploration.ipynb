## <u> Univariate Exploration

### Question 1: What is the distribution of bike riders based on user type?

```python
# absolute frequency of user type 
freq = df_clean['user_type'].value_counts()
gen_order = freq.index
sns.countplot(data=df_clean, x='user_type', color=base_color,
             order=gen_order);
```

```python
# relative frequency of user type
freq = df_clean['user_type'].value_counts()
gen_order = freq.index
ax = sns.countplot(data = df_clean, x = 'user_type', color = base_color, order = gen_order)

total = len(df_clean)
for p in ax.patches:
    tick_names = '{:.2f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()/2
    y = p.get_height()+.05
    ax.annotate(tick_names, (x, y),ha='center')

plt.show();
```

### Observation

According to this visualization, 90.53% of all bike riders are Subscribers, with the Customers only making up the remaining 9.47%.

### Question 2: What is the distribution of bike riders based on gender?

```python
# absolute frequency of member gender
freq = df_clean['member_gender'].value_counts()
gen_order = freq.index
sns.countplot(data=df_clean, x='member_gender', color=base_color,
             order=gen_order);
```

```python
# relative frequency of member gender

freq = df_clean['member_gender'].value_counts()
gen_order = freq.index
ax = sns.countplot(data = df_clean, x = 'member_gender', color = base_color, order = gen_order)

total = len(df_clean)
for p in ax.patches:
    tick_names = '{:.2f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()/2
    y = p.get_height()+.05
    ax.annotate(tick_names, (x, y),ha='center')

plt.show();
```

## <u>Bivariate Exploration

For this section of the exploration, we will mostly be looking the relationships between qualitative and quantitive variables in the dataset. 

Given that we only have one quantitative variable in this dataset i.e duration_sec, we will be plotting this variable against other qualitative variables that are present in the dataset.

### Question 7: Relationship between user related categorical variables i.e member_gender, user_type, and the numerical variable i.e duration_sec

```python
# due to the size of the dataset, it would be difficult to plot in it's entirety which is why we will need to create a 
# variables containing select samples from the dataset.
bike_samples = df_clean.sample(n=2000, replace = False)

# creating a variable to house the member related categorical variables
mem_vars = ['member_gender', 'user_type']

def violinplot(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sns.color_palette()[0]
    sns.violinplot(x=x, y=y, color=default_color, inner = None)

plt.figure(figsize = [10, 10])
g = sns.PairGrid(data = bike_samples, y_vars = ['duration_sec'], x_vars = mem_vars,
                height = 5, aspect = 1.5)
g.map(violinplot)
plt.ylim([0,5000])
plt.show();bike_samples = df_clean.sample(n=2000, replace = False)
```

### Observations

In the first visualization, it would appear that among the three gender categories, the males have the shortest average ride durations, with most of their rides taking less than 1000 seconds.

The females have the second shortest average ride duration, with a slightly more spread out distribution.

While the others seem to the longest average ride duration

In the second visualization, it would appear that subscribers spend a shorter time on average on their rides compared to customers.

### Question 8: Relationship between other categorical variables and the numerical variable

```python
# creating a variable to house the other categorical variables
other_vars = ['start_day_of_week', 'end_day_of_week']

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sns.color_palette()[0]
    sns.boxplot(x=x, y=y, color = base_color, order = ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday']);

plt.figure(figsize = [10, 10])
g = sns.PairGrid(data = bike_samples, y_vars = ['duration_sec'], x_vars = other_vars,
                height = 5, aspect = 1.5)
g.map(boxgrid)
plt.ylim([0,5000])
plt.show();
```

### Observations

The relationship between ride duration and day of week seems to be positive.

##  <u>Multivariate Exploration

In the previous section, we explored the relationship between different qualitative variables present in the dataset and the only quantitave variable present which is duration_sec.

Now in this section, we will explore the relationship between two or more qualitative variables, and duration_sec.

### Question 10: Relationship between member gender, user type and trip duration

```python
g = sns.PairGrid(data = df_clean, y_vars="duration_sec", x_vars=["member_gender", "user_type"], height=5)
g.map(sns.barplot, color = base_color);
g.fig.suptitle('Average trip duration by Gender and User type')
g.fig.subplots_adjust(top=0.88)
g.axes[0,0].yaxis.set_label_text('Trip Duration(secs)')
g.axes[0,0].xaxis.set_label_text('Member gender')
g.axes[0,1].xaxis.set_label_text('User type');
```

### Observations

Although having the lowest count value, 'Other' under the gender type has the highest average trip duration

And, customers have a higher average trip duration than subscribers.
